DATA Filtering & Transformation
:Filtering ✅Sorting ✅GroupBy ✅Aggregation

```python
import pandas as pd

data = {
    "Region": ["South", "North", "East", "West", "South", "North"],
    "Product": ["Laptop", "Mobile", "Laptop", "Tablet", "Mobile", "Tablet"],
    "Sales": [100000, 30000, 45000, 25000, 60000, 46000]
}

df = pd.DataFrame(data)

print(df)
```

```python
North_sales = df[df["Region"]=="North"]

print(North_sales)
```

```python
low_Sales = df[df["Sales"] > 45000]

print(low_Sales)
```

```python
sorted_df = df.sort_values("Sales")

print(sorted_df)
```

```python
region_sales = df.groupby("Region")["Sales"].sum()

print(region_sales)
```

```python
Product_Sales =df. groupby("Product")["Sales"].sum()

print(Product_Sales)
```

```python
print(df["Sales"].sum())
```

```python
print(df["Sales"].mean())
```

```python
print(df["Sales"].max() )
```

```python
print(df["Sales"].min())
```

```python
print(df["Sales"].count())
```

```python
result =df.groupby("Region")["Sales"].agg(
    ["sum" ,"mean","max","min"]

)
print(result)
```

```python
import matplotlib.pyplot as plt


months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales = [10000, 12000, 15000, 13000, 18000, 20000]


plt.figure(figsize=(10,5))


plt.subplot(1, 3, 1)
plt.plot(months, sales, marker="o")
plt.title("Sales Trend")
plt.xlabel("Months")
plt.ylabel("Sales")


plt.subplot(1, 3, 2)
plt.bar(months, sales, color="green")



plt.subplot(1,3,3)

expenses = [5000,3000,2000]
categories = ["Food","Travel","Shopping"]

plt.pie(expenses,
        labels=categories,
        autopct="%1.1f%%", colors = ["pink","yellow","green"])

plt.title("Expense Distribution")


plt.tight_layout()
plt.show()
```
