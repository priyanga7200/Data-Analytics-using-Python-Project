```python
def attendance_peracentage(attended,total):
  percentage=(attended/total)*100
  return round(percentage,2)

def check_eligibilty(percent):
  if percent>=75:
    return "ELIGIBLE"
  else:
    return"NOT ELIGIBLE X"

try:
  total = int(input('Total classes:'))
  attended = int(input('classes Attended:'))

  pct = attendance_peracentage(attended,total)
  status = check_eligibilty(pct)


  print( f'Attendance:{pct}%')
  print(f'Status:{status}')

except ValueError:
  print('Enter valid numbers!')
```
