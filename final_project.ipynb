# **TOTAL** **AVERAGE**

```python
import pandas as pd

df["Total_Average"] = df[["Maths", "Science", "English"]].mean(axis=1)

print(df[["Name", "Total_Average"]])
```

# **GRADE**

```python
def grade(x):
    if x >= 90:
        return "A"
    elif x >= 75:
        return "B"
    elif x >= 50:
        return "C"
    else:
        return "D"

df["Total_Average"] = df[["Maths", "Science", "English"]].mean(axis=1)
df["Grade"] = df["Total_Average"].apply(grade)

print(df[["Name", "Total_Average", "Grade"]])
```

# **TOPPER**

```python
df["Total_Average"] = df[["Maths", "Science", "English"]].mean(axis=1)

topper = df.loc[df["Total_Average"].idxmax()]

print("🏆 TOPPER DETAILS")
print(topper)
```

# **WEAK** **STUDENTS**

```python
df["Total_Average"] = df[["Maths", "Science", "English"]].mean(axis=1)

weak_students = df[df["Total_Average"] < 50]

print("⚠️ WEAK STUDENTS")
print(weak_students)
```

# **SUBJECT**-**WISE** **PERFORMANCE**

```python
import matplotlib.pyplot as plt

subject_avg = df[["Maths", "Science", "English"]].mean()

colors = ["red", "green", "blue"]

subject_avg.plot(kind="bar", color=colors)

plt.title("Subject-wise Performance")
plt.ylabel("Average Marks")
plt.show()
```

# **RESULT**

```python
df["Result"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Pass vs Fail")
plt.show()
```

## **GRADE** **DISTRIBUTION**

```python
df["Grade"].value_counts().plot(kind ="bar",
                                color ="purple")
plt.title("Grade Distribution")
plt.ylabel("Number of students")
plt.show()
```
