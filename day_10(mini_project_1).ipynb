```python
df =df.drop_duplicates()

print(df)
```

```python
total_sales = df["Total_Sales"].sum()

print("Total Sales:", total_sales)
```

```python
Product_Sales =df.groupby("Product")["Total_Sales"].sum()

print(Product_Sales)
```

```python
Product_Sales.plot(kind ='bar')

plt.title("Product Sales Analysis")

plt.xlabel("Product")

plt.ylabel("Sales")
```
