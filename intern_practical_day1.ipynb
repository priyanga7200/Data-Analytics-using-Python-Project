```python
mark1=int(input("Enter marks for subject 1:"))
mark2=int(input("Enter marks for subject 2:"))
mark3=int(input("Enter marks for subject 3:"))

total=mark1+ mark2+ mark3
Average=total/3

print("Total marks=",total)
print("Average marks=",Average)

if Average >=50:
  print("pass")
else:
  print("Fail")


```

```python
def get_grade(mark):
  if mark >=90:
   return 'S'
  elif mark >=75:
    return 'A'
  else:
    return 'F'

student ={
    "Deva":95,
    "Priya":82,
    "Arun":65,
    "Diya":91
}
passed = 0
failed = 0

for name,mark in student.items():
     grade = get_grade(mark)
     print(name, "_", mark, "_", grade)


     if grade == 'F':
        failed +=1
     else:
       passed +=1

     print("Passed:",passed)
     print("Failed:",failed)
```
